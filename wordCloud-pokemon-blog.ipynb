```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from wordcloud import WordCloud, ImageColorGenerator
```

```python
text = open('pokemon-names.txt').read()
orig_image = np.array(Image.open("pikachu.png"))

# create coloring from image
image_colors = ImageColorGenerator(orig_image)

wc = WordCloud(background_color="white", max_words=1000, repeat=True,
               mask=orig_image, 
               color_func=image_colors,
               stopwords=set(),
               max_font_size=40, min_font_size=3, random_state=11013)
wc.generate(text)
```

```python
# show
fig, axes = plt.subplots(1, 2)
fig.set_size_inches((20,25))

axes[0].imshow(wc, interpolation="bilinear")
axes[1].imshow(orig_image, interpolation="bilinear")
for ax in axes:
    ax.set_axis_off()
    
# must save before calling plt.show()    
#plt.savefig('pokemon-comp-2.png',bbox_inches='tight')

plt.show()
```
